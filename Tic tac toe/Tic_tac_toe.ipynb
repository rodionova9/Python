```python
import canvas, time, random

canvas.set_color('green')
canvas.line_width(4)
canvas.stroke_rect(0, 0, 350, 350)
canvas.stroke_rect(116, 0, 116, 350)
canvas.stroke_rect(0, 116,  350, 116)

GAME_STATE = [None, None, None, None, None, None, None, None, None]


def draw_state(GAME_STATE): # кр нол
    slow =  ['58:58', '174:58', '290:58', '58:174', '174:174', '290:174', '58:290', '174:290', '290:290']    

    for z, i in enumerate(slow):
        c=i.find(':')
        s=int(i[:c])
        j=int(i[(c+1):])
        #print(s, j)

        if GAME_STATE[z]!=None:
            if GAME_STATE[z]=="x":
                canvas.set_color('blue') 
                canvas.radius_line(s-30, j+30, 45, 80)
                canvas.radius_line(s+30, j+30, 315 , 80)

            if GAME_STATE[z]=="o":
                canvas.set_color('green')     
                canvas.circle(s, j, 30)        

    canvas.draw()

def get_bot_move (GAME_STATE): #бот
    b=[]

    for i, d in enumerate(GAME_STATE):
        if d==None:
            b.append(i)

    q=random.choice(b)

    return q

def get_winner(GAME_STATE): 

    r=['012', '345', '678', '036', '147', '258', '048', '246']

    if None  not in  GAME_STATE:

        for i, e in enumerate(r):  
            y=int(e[0:1])
            t=int(e[1:2])
            u=int(e[2:3])

            if GAME_STATE[y]=="x" and GAME_STATE[t]=="x" and  GAME_STATE[u]=="x" :
                v="x_win"
                return(v)


            if GAME_STATE[y]=="o" and  GAME_STATE[t]=="o" and  GAME_STATE[u]=="o" :
                v="o_win"
                return(v)

        else:
            v="draw"
            return(v)

    else:
        for i, e in enumerate(r):  
            y=int(e[0:1])
            t=int(e[1:2])
            u=int(e[2:3])

            if GAME_STATE[y]=="x" and GAME_STATE[t]=="x" and  GAME_STATE[u]=="x" :
                v="x_win"
                return(v)


            if GAME_STATE[y]=="o" and  GAME_STATE[t]=="o" and  GAME_STATE[u]=="o" :
                v="o_win"
                return(v)

        else:
            v=None
            return(v)



print(get_winner(GAME_STATE)) 


def click(x, y): # клик 
    global GAME_STATE 
    for k in range(0, 8, 1) :
        f=x//116
        g=y//116
        k= f+ 3*g 

        if GAME_STATE[k]==None:
            GAME_STATE[k]="x" 
            try:
                s=get_bot_move (GAME_STATE)

            except:
                pass

            else:    
                if GAME_STATE[s]==None and GAME_STATE[s]!="x":
                    GAME_STATE[s]="o"
    print (GAME_STATE)
    draw_state(GAME_STATE)  
    print(get_winner(GAME_STATE))   
draw_state(GAME_STATE)

canvas.set_onclick(click)
canvas.listen() 
```
